### Linear Regression

```python
## Splitting the data into train and test data set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 4)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
print("accuracy is {} on training data".format(linear_model.score(x_train, y_train)))
```

```python
print("accuracy is {} on test data".format(linear_model.score(x_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score
model = cross_val_score(linear_model, x_train, y_train, cv = 5)
```

```python
model.mean()
```
